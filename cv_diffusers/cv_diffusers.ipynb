```python
from diffusers import StableDiffusionPipeline
import torch
from huggingface_hub import login
import time

login()
```

```python
from google.colab import drive

drive.mount('/content/drive/')

def backup_colab_content_to_drive(folder_name='Colab Notebooks'):
  import shutil
  import os

  src = '/content'
  dest = f'/content/drive/MyDrive/{folder_name}'
  os.makedirs(dest, exist_ok=True)

  for item in os.listdir(src):
    if item == 'drive':
      continue
    s = os.path.join(src, item)
    d = os.path.join(dest, item)
    if os.path.isdir(s):
      shutil.copytree(s, d)
    else:
      shutil.copy2(s, d)

  print(f'Backup complete. Files saved to: {dest}')
```

```python
pipe_stable = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4"
)
```

```python
prompts = {
    'prompt_simple': 'Generate an Ukrainian writer.',
    'prompt_detailed': 'Generate an Ukrainian writer in the garden with Ukrainian landscape.',
    'prompt_creative': 'Generate a futuristic Ukrainian writer in the futuristic garden.'
}
```

```python
def generate_image(pipe, prompt_name, prompt):
  start = time.time()
  image = pipe(prompt).images[0]
  image.save(f"image_{pipe.name_or_path.replace('/', '_')}_{prompt_name}.png")
  print (f'Model {pipe} generated {prompt_name} image in {time.time()-start}')


for pr in prompts.keys():
  generate_image(pipe=pipe_stable, prompt_name=pr, prompt=prompts[pr])
```
